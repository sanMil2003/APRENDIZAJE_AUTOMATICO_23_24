La elección de funciones de activación en modelos de inteligencia artificial, como redes neuronales, depende de la naturaleza del problema que estás abordando y de las características de los datos. La función de activación determina la salida de una neurona, y hay varias funciones comúnmente utilizadas, entre ellas ReLU (Rectified Linear Unit) y sigmoid.

ReLU (Rectified Linear Unit):

La función ReLU es f(x) = max(0, x). Es simple y computacionalmente eficiente.
Su salida es 0 para valores negativos y lineal para valores positivos, lo que facilita el aprendizaje y la convergencia rápida en muchos casos.
Es especialmente útil en problemas de clasificación, como el reconocimiento de imágenes.
Sigmoid:

La función sigmoide tiene la forma f(x) = 1 / (1 + e^(-x)).
Produce valores en el rango (0, 1), lo que es útil en problemas de clasificación binaria, ya que puede interpretarse como la probabilidad de pertenencia a la clase positiva.
Sin embargo, la función sigmoide puede sufrir del problema de desvanecimiento del gradiente durante el entrenamiento de redes neuronales profundas.
Diferencias:

Rango de Salida:

ReLU produce salidas en el rango [0, ∞), mientras que sigmoid produce salidas en el rango (0, 1). La salida de ReLU no está acotada en la parte superior.
Problema de Desvanecimiento del Gradiente:

Sigmoid puede sufrir del problema de desvanecimiento del gradiente en redes neuronales profundas, donde los gradientes se vuelven muy pequeños durante la retropropagación. ReLU tiende a mitigar este problema.
Sparsity:

ReLU induce esparsidad ya que muchas de sus activaciones son cero. Esto puede ser beneficioso en términos de eficiencia computacional y puede actuar como una forma de regularización.
Sigmoid no induce naturalmente esparsidad.
Problema de Activaciones Muertas:

ReLU puede enfrentar el problema de activaciones muertas, donde algunas neuronas pueden quedar inactivas para siempre si su salida es siempre cero para todas las entradas durante el entrenamiento.
Sigmoid no tiene este problema, pero puede saturarse fácilmente para entradas extremas.

```python
# a mano inicializar pesos
#xabier
def init_xavier(neurons):
  weights = {}
  for i in range(len(neurons)-1):
    weights['L' + str(i+1)] = np.ramdom.randn(neurons[1+i], neurons[i] / np.sqrt(neurons[i]))
  return weights

# kaiming He -- mejor para Relu
def init_kaiming(neurons):
  weights = {}
  for i in range(len(neurons)-1):
    weights['L' + str(i+1)] = np.ramdom.randn(neurons[1+i], neurons[i]) / np.sqrt(neurons[i]/2)
  return weights
```


La inicialización de Kaiming o He es especialmente adecuada para funciones de activación rectificadas, como ReLU (Rectified Linear Unit). La razón principal es que estas funciones introducen no linealidades en el modelo al activar ciertas neuronas y desactivar otras. Sin embargo, estas funciones también pueden introducir problemas durante la inicialización de los pesos.

Cuando inicializas los pesos de una red neuronal, es importante que la magnitud de la entrada no cause una explosión o desvanecimiento del gradiente. En el caso de funciones de activación lineales o sigmoidales, las funciones de inicialización comunes, como Glorot o Xavier, son adecuadas. Sin embargo, las funciones de activación rectificadas, como ReLU, pueden introducir un sesgo hacia la desvanecimiento del gradiente si no se inicializan adecuadamente.

La función de activación ReLU establece todas las entradas negativas a cero. Si los pesos están inicializados con una varianza demasiado alta, puede suceder que muchas de las neuronas se activen y otras se desactiven, lo que resulta en una salida esparsa y una propagación ineficiente de los gradientes.

La inicialización de Kaiming o He aborda este problema ajustando la varianza de los pesos según la cantidad de entradas que tiene cada neurona. Esto ayuda a mantener una propagación de gradientes más equilibrada y evita que la mayoría de las neuronas se activen o desactiven al comienzo del entrenamiento.

```python
# 3.- CONSTRUIMOS LA ARQUITECTURA DE LA RED
# es mejor relu para capas oculatas y sigmoid para capas de salida
from keras import models
from keras import layers

# inicilizar pesos de forma automática
# GlorotUniform = inicialización de Xavier
# con Kaiming sería: from keras.initializers import HeNormal
## from keras.initializers import GlorotUniform, RandomNormal
# inicializamos los pesos con distribución normal
## initializer_first_layer = GlorotUniform()
## initializer_second_layer = RandomNormal(mean=0.0, stddev=0.01)
#creamos modelo
## network = models.Sequential()
## network.add(layers.Dense(512, activation='sigmoid', input_shape=(28*28,),
##                          kernel_initializer = initializer_first_layer))
## network.add(layers.Dense(10, activation='softmax',
##                          kernel_initializer = initializer_second_layer))

# El tipo de red será secuencial. Desde la entrada hasta la salida sin ciclos
network = models.Sequential()
# Creamos dos capas "Dense", que son capas neuronales densamente conectadas
# (también llamadas "completamente conectadas"). Cada una de las 512 neuronas de
# la capa de entrada están conectadas con los 784 píxeles = 28*28. Solo lo
# definimos para la primera capa. Para la segunda capa y posteriores, Keras lo
# deduce.
## network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(512, activation='sigmoid', input_shape=(28*28,)))
# Probar 10 neuronas y 'sigmoid' <==============================================
# Capa de salida "softmax" de 10 vías (o neuronas). Significa que
# devolverá una matriz de 10 puntuaciones de probabilidad (sumando 1)
# La puntuación será la probabilidad de que la imagen del dígito actual
# pertenezca a una de nuestras clases de 10 dígitos.
network.add(layers.Dense(10, activation='softmax'))
# Cada capa aplica unas cuantas operaciones con tensores sencillas a los datos
# de entrada, y que estas operaciones implican tensores de pesos. Los tensores
# de peso, que son los atributos de las capas, son donde persiste el
# "conocimiento" de la red.
# En general, la capa de salida de una red de clasificación tendrá tantas
# neuronas como clases, menos en la clasificación binaria, que con 1 vale. Cada
# valor será la probabilidad de que la imagen del dígito actual pertenezca a cada
# una de las clases
network.summary()
# Nombre de las capas automáticos a no ser que lo definamos
# 401.920 = 784 x 512 + 512 Sesgo
# 5.130 = 512x10 + 10 Sesgo
# 407.050 = 401.920 + 5.130
```

**SOFTMAX**

* La función **softmax** en la que la salida de cada neurona corresponde a la probabilidad estimada de la clase correspondiente. Dada una imagen de entrada, obtendremos las probabilidades de que sea cada uno de los 10 posibles dígitos. Esta aproximación de usar una distribución de probabilidades nos puede dar una mejor idea de cuán confiados estamos de nuestra predicción.

* Para cada ejemplo de entrada obtendremos como vector de salida de la red neuronal una distribución de probabilidad sobre un conjunto de etiquetas mutuamente excluyentes, es decir, un vector de 10 probabilidades cada una correspondiente a un dígito y que todas estas 10 probabilidades sumen 1 (las probabilidades se expresarán entre 0 y 1).

* La función softmax se basa en calcular **“las evidencias”** de que una determinada imagen pertenece a una clase en particular y luego se convierten estas evidencias en probabilidades de que pertenezca a cada una de las posibles clases

* Para medir la **evidencia** de que una determinada imagen pertenece a una clase en particular, una aproximación consiste en realizar una **suma ponderada de la evidencia de pertenencia** de **cada uno de sus píxeles** a esa clase.

* Ejemplo con el 0: 28×28 píxeles, los píxeles en rojo  representa pesos negativos (es decir, reducir la evidencia de que pertenece), mientras que los píxeles en azul representan pesos positivos (aumenta la evidencia de que pertenece). El color negro representa el valor neutro

SGD (Stochastic Gradient Descent):

SGD es uno de los optimizadores más básicos y ampliamente utilizados.
Actualiza los parámetros en la dirección opuesta al gradiente, escalado por una tasa de aprendizaje constante.
Puede ser sensible a la elección de la tasa de aprendizaje y puede tener dificultades en converger en problemas con geometría complicada de la función de pérdida.
RMSprop (Root Mean Square Propagation):

RMSprop es un optimizador que adapta la tasa de aprendizaje para cada parámetro individualmente.
Introduce una media móvil ponderada de los cuadrados de los gradientes anteriores en el cálculo de la tasa de aprendizaje. Esto ayuda a atenuar la sensibilidad a la elección de la tasa de aprendizaje y puede acelerar la convergencia en problemas difíciles.
Diferencias y Consideraciones:

RMSprop a menudo es preferido en problemas donde la escala de los datos puede variar significativamente entre diferentes dimensiones o características, ya que adapta la tasa de aprendizaje en consecuencia.
En situaciones donde la función de pérdida tiene regiones planas o características muy diferentes en diferentes direcciones, RMSprop puede ayudar a converger más rápidamente que SGD, que usa la misma tasa de aprendizaje para todas las direcciones.
La elección entre RMSprop y SGD puede depender del problema específico, la arquitectura de la red y la configuración de hiperparámetros.

Veamos la matriz de confusión de nuestro ejemplo:

```python
# Look at confusion matrix
# Note, this code is taken straight from the SKLEARN website, an nice way of
# viewing confusion matrix.
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Observación')
    plt.xlabel('Predicción')
```

```python
from collections import Counter
from sklearn.metrics import confusion_matrix
import itertools
import numpy as np

# Predicción de valores desde el dataset de validación
Y_pred = network.predict(test_images)
print("=VALORES QUE PREDICE EL MODELO ENTRENADO CON LAS IMAGENES DE TEST======")
print("=RECORDAR QUE LA PRIMERA IMAGEN DE TEST ERA UN 7=======================")
print(Y_pred[0])
print("=======================================================================")
# Convertir en predicciones de clases desde los vectores one-hot encoding
Y_pred_classes = np.argmax(Y_pred, axis = 1)
print("=CONVERSION DE PREDICCIONES EN CLASES DESDE VECTORES ONE-HOT ENCODING==")
print(Y_pred_classes)
print("=======================================================================")
# Convertir en clases de observación desde los vectores one-hot encoding
Y_true = np.argmax(test_labels, axis = 1)
print("=CONVERSION DE VALORES REALES EN CLASES DESDE VECTORES ONE-HOT ENCODING")
print(Y_true)
print("=======================================================================")
# Computar la matriz de confusión
confusion_mtx = confusion_matrix(Y_true, Y_pred_classes)
print("=MATRIZ DE CONFUSION CON VALORES REALES FRENTE A LOS PREDICHOS=========")
print(confusion_mtx)
print("=======================================================================")
# Dibujar la matriz de confusión
print("=PINTAMOS LA MATRIZ DE CONFUSION=======================================")
plot_confusion_matrix(confusion_mtx, classes = range(10))
```

## Distribución de Probalidades

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label.all() == true_label.all():
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(predicted_label,
                                100*np.max(predictions_array),
                                true_label),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#00FF00")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('black')
```

La elección entre ReLU (Rectified Linear Unit) y Sigmoid depende en gran medida del problema específico que estás abordando y las características de tus datos. Aquí hay algunas consideraciones para cada función de activación:

ReLU (Rectified Linear Unit):
No satura para entradas positivas: La principal ventaja de ReLU es que no satura para entradas positivas. Esto significa que no "mata" los gradientes durante la retropropagación cuando la entrada es positiva, lo que ayuda a mitigar el problema del desvanecimiento del gradiente en redes profundas.

Computacionalmente eficiente: La función ReLU es computacionalmente eficiente, ya que simplemente apaga todas las entradas negativas a cero.

Favorece la sparse activation: Al establecer las entradas negativas a cero, ReLU tiende a activar solo una fracción de las neuronas, lo que conduce a una representación más dispersa y puede ayudar a aprender representaciones más útiles.

Sigmoid:
Limitado entre 0 y 1: La función Sigmoid es útil cuando necesitas que la salida de la neurona esté limitada entre 0 y 1. Esto es especialmente común en la capa de salida de modelos de clasificación binaria, donde la salida puede interpretarse como la probabilidad de pertenecer a la clase positiva.

Suavidad y derivadas bien definidas: La función Sigmoid tiene derivadas bien definidas en todos los puntos, lo que facilita la propagación del gradiente durante el entrenamiento. Sin embargo, puede sufrir el problema de saturación de gradientes para entradas muy grandes o muy pequeñas.

En la práctica, ReLU se ha vuelto muy popular en las capas ocultas de las redes neuronales profundas debido a su eficiencia y capacidad para mitigar problemas de desvanecimiento del gradiente. Sin embargo, en la capa de salida de una red para clasificación binaria, la función Sigmoid sigue siendo común.

En muchos casos, las funciones de activación como ReLU, Leaky ReLU o variantes de ellas (como Parametric ReLU o Exponential Linear Unit) se prefieren en capas ocultas debido a su capacidad para manejar mejor los problemas de desvanecimiento del gradiente y su eficiencia computacional. Es común usar Sigmoid en la capa de salida para problemas de clasificación binaria, y Softmax para problemas de clasificación multiclase.

# RETOMAMOS EL MODELO ORIGINAL PARA PINTAR LOS PESOS APRENDIDOS

```python
%tensorflow_version 2.x
import tensorflow as tf
from tensorflow import keras
print(tf.keras.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))
print(tf.__version__)
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
from keras import models
from keras import layers
network = models.Sequential()
network.add(layers.Dense(512, name='Capa_Entrada_Relu', activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, name='Capa_Salida_Softmax', activation='softmax'))
network.summary()
print("=======================================================================")
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
from keras.utils.np_utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
network.fit(train_images, train_labels, epochs=5, batch_size=128)
print("=======================================================================")
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_loss:', test_loss)
print('test_acc:', test_acc)
print("=======================================================================")
Y_pred = network.predict(test_images)
print('Predicción para la primera imagen de test:', Y_pred[0])
print("=======================================================================")
Y_pred_classes = np.argmax(Y_pred, axis = 1)
print("Valor real para la primera imagen de test:", Y_pred_classes[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
w = network.get_weights()
# Capa 0 => Pesos
first_layer_weights = network.layers[0].get_weights()[0]
print('Forma de los Pesos de la capa 0: ',first_layer_weights.shape)
# Capa 0 => Umbrales
first_layer_biases  = network.layers[0].get_weights()[1]
print('Forma de los Umbrales de la capa 0: ',first_layer_biases.shape)
# Capa 1 => Pesos
second_layer_weights = network.layers[1].get_weights()[0]
print('Forma de los Pesos de la capa 1: ',second_layer_weights.shape)
# Capa 1 => Umbrales
second_layer_biases  = network.layers[1].get_weights()[1]
print('Forma de los Umbrales de la capa 1: ',second_layer_biases.shape)
```
